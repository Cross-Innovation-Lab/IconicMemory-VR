### XGBOOST模型比较

```python
import shap
from xgboost import XGBRegressor
from sklearn.metrics import f1_score, r2_score,mean_squared_error,mean_absolute_error,accuracy_score,precision_score, recall_score
from Realize_0312 import states,emotions
from interfaces import ML_record
from sklearn.metrics import mean_squared_error #均方误差
from sklearn.metrics import mean_absolute_error #平方绝对误差
shap.initjs()
import numpy as np

test_0315=ML_record(attribute_name=['max_depth','n_estimators'])
depth=[5,10,15,20,25,30,35,40,45,50]
n_estimators=[3,5,15,20,25]
for state in states[1:]:
    for dep in depth:
        for esti in n_estimators:
            dft,dfe=f.datasetDivide_BySubject(dataset_name=state+'_r_time')
            predict='r_time'
            X=dft.drop(predict,axis=1)
            y=dft[predict]
            model = XGBRegressor( n_estimators = esti,
                               learning_rate = 0.01, #学习率
                               max_depth=dep)
            model.fit(X, y)
            X2=dfe.drop(predict,axis=1)
            y2=dfe[predict]
            ypred = model.predict(X2)
            test_0315.rec_predict(dataset=state+'_r_time',atr={'max_depth':dep,'n_estimators':esti},y=y2,pre_y=ypred)
test_0315.save_csv_r_time('XGboost_r_time')
# explainer = shap.TreeExplainer(model)
# shap_values=explainer.shap_values(X)
# shap.force_plot(explainer.expected_value, shap_values[0,:], X.iloc[0,:])
# shap.summary_plot(shap_values, X)
```

### SVM模型比较

```python
import shap
from sklearn.svm import SVR
from sklearn.metrics import f1_score, r2_score,mean_squared_error,mean_absolute_error,accuracy_score,precision_score, recall_score
from Realize_0312 import states,emotions
from interfaces import ML_record
from sklearn.metrics import mean_squared_error #均方误差
from sklearn.metrics import mean_absolute_error #平方绝对误差

recorder=ML_record(attribute_name=['c','kernal'])
kernals=['linear', 'poly', 'rbf', 'sigmoid']
for state in states[1:]:
       for c in range(10,20,2):
        for ker in kernals:
            dft,dfe=f.datasetDivide_BySubject(dataset_name=state+'_r_time')
            predict='r_time'
            X=dft.drop(predict,axis=1)
            y=dft[predict]
            model = SVR(C=float(c)/20.0,kernel=ker)

            model.fit(X, y)
            X2=dfe.drop(predict,axis=1)
            y2=dfe[predict]
            ypred = model.predict(X2)
            recorder.rec_predict(dataset=state+'_r_time',atr={'c':float(c)/20.0,'kernal':ker},y=y2,pre_y=ypred)
recorder.save_csv_r_time('SVM_r_time')
```

```python
import shap
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import f1_score, r2_score,mean_squared_error,mean_absolute_error,accuracy_score,precision_score, recall_score
from Realize_0312 import states,emotions
from interfaces import ML_record
from sklearn.metrics import mean_squared_error #均方误差
from sklearn.metrics import mean_absolute_error #平方绝对误差

atr_name=['max_depth','criterion']
criterion=['squared_error','absolute_error',"poisson",'friedman_mse']
predict='r_time'
recorder=ML_record(attribute_name=atr_name)
for state in states[1:]:
   for cri in criterion:
        for dep in range (5,46,5):
            model = DecisionTreeRegressor(criterion=cri,max_depth=dep,random_state=7)# 实例化模型，添加criterion参数
            dft,dfe=f.datasetDivide_BySubject(dataset_name=state+'_r_time')

            X=dft.drop(predict,axis=1)
            y=dft[predict]

            model.fit(X, y)
            X2=dfe.drop(predict,axis=1)
            y2=dfe[predict]
            ypred = model.predict(X2)
            recorder.rec_predict(dataset=state+'_r_time',atr={'max_depth':dep,'criterion':cri},y=y2,pre_y=ypred)
test_0315.save_csv_r_time('Regression_tree_r_time')
```
